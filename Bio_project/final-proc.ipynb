# Get the json file, and its corresponding label files

```python
# All imports
import os
import json
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

Identify files that have their labels as well, print out the ones that don't have their corresponding label files

```python
i = 0
list_of_files = []
for f in os.listdir("./AGjsons/"):
    if os.path.isfile("./labels/" + str(f)):
        list_of_files.append(f)
#         print f, "exists along with its label file"
        i += 1
    else:
        print f
print "\n", i, "is the number of files with their label files"
```

Through each file, extract the contents and update the matrix X and Y. Matrix X has the data of each file laid out in rows

```python
taxons = None
X = []
Y = []

i = 0
for f in list_of_files:
    print f
    x_dic = json.loads(open("./AGjsons/" + f).read())
    
    # Just one time, read all the taxons
    if i == 0:
        taxons = sorted(x_dic.keys())
    
    # Append X to have the data
    X.append([x_dic[k] for k in taxons])
    
    y_dic = json.loads(open("./labels/" + f).read())
    
    # Append Y to have the label for that file
    if y_dic['sex'] == "female":
        Y.append(1)
    else:
        Y.append(0)

```

```python
X = np.array(X)
Y = np.array(Y)

# To confirm the lengths of the data
print np.shape(X)
print len(Y)
```

```python
# Check X
X[0, :]
```

## Filter the data by deleting taxons that have a variance which lies outside a range

```python
# Set the range for the tolerance on variances
range_low = 50
range_high = 150
```

Go through every column of the data, get the variance. If the variance is out of range, remove that column

```python
tmp_X = X
taxons_to_keep = []
cols_to_keep = []
for i in range(np.shape(tmp_X)[1]):
    var = np.var(tmp_X[:, i])
    if var > range_low and var < range_high:
        taxons_to_keep.append(taxons[i])
        cols_to_keep.append(i)
#     print "col", i, "= ", var
```

### The taxons that passed our variance test

```python
print taxons_to_keep
```

### New X generated after only keeping the taxons that passed our test

```python
new_X = np.ones((np.shape(X)[0], len(taxons_to_keep)))
print np.shape(new_X)
for i in range(len(cols_to_keep)):
    new_X[:, i] = X[:, cols_to_keep[i]]
    
print new_X
print len(Y)
```

# Run the neural net on the processed data

```python
X_train, X_test, y_train, y_test = train_test_split(new_X, Y, test_size=0.20, shuffle=False)

# Specify feature
feature_columns = [tf.feature_column.numeric_column("x", shape=[np.shape(X_train)[1]])]
```

```python
# Build 2 layer DNN classifier
# classifier = tf.estimator.DNNClassifier(
#     feature_columns=feature_columns,
#     hidden_units=[20, 5],
#     optimizer=tf.train.ProximalAdagradOptimizer(
#       learning_rate=0.1,
#       l1_regularization_strength=0.001
#     ),
#     n_classes=2,
#     dropout=0.5,
#     model_dir="./tmp/mnist_model"
# )

classifier = tf.estimator.DNNClassifier(feature_columns=feature_columns, hidden_units=[50,50],n_classes=2, dropout=0.5, optimizer=tf.train.ProximalAdagradOptimizer(
      learning_rate=0.1,
      l1_regularization_strength=0.001
    ), batch_norm=True)

# Define the training inputs
train_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"x": X_train},
    y=y_train,
    num_epochs=None,
    batch_size=128,
    shuffle=True
)

classifier.train(input_fn=train_input_fn, steps=100000)
```

```python
# Define the test inputs
test_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"x": X_test},
    y=y_test,
    num_epochs=1,
    shuffle=False
)

# Evaluate accuracy
accuracy_score = classifier.evaluate(input_fn=test_input_fn)["accuracy"]

print("\nTest Accuracy: {0:f}%\n".format(accuracy_score*100))
```
